```
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Read the CSV files
confirmed_cases_df = pd.read_csv("time_series_covid19_confirmed_US.csv")
deaths_df = pd.read_csv("time_series_covid19_deaths_US.csv")

# Select essential columns (using .iloc to get all date columns dynamically)
confirmed_cases_df = confirmed_cases_df.iloc[:, :5].join(confirmed_cases_df.iloc[:, 11:])
deaths_df = deaths_df.iloc[:, :5].join(deaths_df.iloc[:, 12:])

# Choose specific time series columns for analysis
county = "Los Angeles"  # Example county
county_confirmed_cases = confirmed_cases_df[confirmed_cases_df["FIPS"] == county].filter(like="/").sum()

# Ensure county_confirmed_cases is not constant
if county_confirmed_cases.nunique() > 1:
    # Perform stationarity check (Dickey-Fuller test)
    result = adfuller(county_confirmed_cases)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    for key, value in result[4].items():
        print('Critical Values:')
        print(f'   {key}, {value}')
else:
    print("The time series data is constant or has insufficient variation.")
```

```
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

# Function to perform Dickey-Fuller test
def adf_test(timeseries):
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    return dfoutput

# Example usage:
adf_test_results = adf_test(df['UID'])
print(adf_test_results)

# If series is non-stationary (p-value > 0.05), apply differencing
if adf_test_results['p-value'] > 0.05:
    df['stationary_value'] = df['value'].diff().dropna()
else:
    df['stationary_value'] = df['UID']

# Optionally, plot the original and differenced series
plt.figure(figsize=(12, 6))
plt.plot(df['UID'], label='Original Series')
plt.plot(df['stationary_value'], label='Stationary Series')
plt.legend()
plt.show()
```

```
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# ACF and PACF plots
plot_acf(df['stationary_value'], lags=20)
plt.show()

plot_pacf(df['stationary_value'], lags=20)
plt.show()
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_absolute_percentage_error

# Read the CSV files
confirmed_cases_df = pd.read_csv("time_series_covid19_confirmed_US.csv")
deaths_df = pd.read_csv("time_series_covid19_deaths_US.csv")

# Select essential columns (using .iloc to get all date columns dynamically)
confirmed_cases_df = confirmed_cases_df.iloc[:, :5].join(confirmed_cases_df.iloc[:, 11:])
deaths_df = deaths_df.iloc[:, :5].join(deaths_df.iloc[:, 12:])

# Handle missing values
confirmed_cases_df.fillna(0, inplace=True)
deaths_df.fillna(0, inplace=True)

# Choose specific time series columns for analysis
county = "Los Angeles"  # Example county
county_confirmed_cases = confirmed_cases_df[confirmed_cases_df["FIPS"] == county].filter(like="/").sum()

# Prepare data for LSTM (scaling and converting timestamps)
values = county_confirmed_cases.values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_values = scaler.fit_transform(values)

# Convert series to supervised learning
def create_dataset(data, look_back=1):
    X, y = [], []
    for i in range(len(data) - look_back - 1):
        X.append(data[i:(i + look_back), 0])
        y.append(data[i + look_back, 0])
    return np.array(X), np.array(y)

look_back = 30  # Number of previous time steps to use as input variables
X, y = create_dataset(scaled_values, look_back)

# Reshape input to be [samples, time steps, features]
X = np.reshape(X, (X.shape[0], look_back, 1))

# Split into train and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Define LSTM model architecture
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(look_back, 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Train LSTM model on the training set
model.fit(X_train, y_train, epochs=20, batch_size=1, verbose=2)

# Make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Invert predictions
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train = scaler.inverse_transform([y_train])
y_test = scaler.inverse_transform([y_test])

# Evaluate model performance (MAPE)
train_mape = mean_absolute_percentage_error(y_train[0], train_predict[:, 0])
test_mape = mean_absolute_percentage_error(y_test[0], test_predict[:, 0])
print(f'Train MAPE: {train_mape:.4f}')
print(f'Test MAPE: {test_mape:.4f}')

# Plot the results
plt.figure(figsize=(14, 7))
plt.plot(county_confirmed_cases.index[look_back + 1:train_size + look_back + 1], y_train[0], label='Train Data')
plt.plot(county_confirmed_cases.index[train_size + look_back + 1:], y_test[0], label='Test Data')
plt.plot(county_confirmed_cases.index[look_back + 1:train_size + look_back + 1], train_predict[:, 0], label='Train Predict')
plt.plot(county_confirmed_cases.index[train_size + look_back + 1:], test_predict[:, 0], label='Test Predict')
plt.title(f'LSTM Model Forecast for COVID-19 Confirmed Cases in {county} County')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# Normalize data
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(np.array(train).reshape(-1, 1))
test_scaled = scaler.transform(np.array(test).reshape(-1, 1))

# Function to prepare data for LSTM
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X[i:(i + time_steps)]
        Xs.append(v)
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

# Prepare data with time steps
time_steps = 10
X_train, y_train = create_dataset(train_scaled, train_scaled, time_steps)
X_test, y_test = create_dataset(test_scaled, test_scaled, time_steps)

# Build LSTM model
model = Sequential([
    LSTM(units=50, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')

# Train LSTM model
model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=1)

# Predict on test data
lstm_forecast_scaled = model.predict(X_test)
lstm_forecast = scaler.inverse_transform(lstm_forecast_scaled).flatten()

# Calculate MAPE for LSTM forecast
lstm_mape = mean_absolute_percentage_error(test.values[time_steps:], lstm_forecast)
print(f'LSTM Mean Absolute Percentage Error (MAPE): {lstm_mape}')

# Optionally, plot actual vs. forecasted values
plt.figure(figsize=(12, 6))
plt.plot(test.index[time_steps:], test.values[time_steps:], label='Actual')
plt.plot(test.index[time_steps:], lstm_forecast, label='LSTM Forecast')
plt.legend()
plt.show()
```
